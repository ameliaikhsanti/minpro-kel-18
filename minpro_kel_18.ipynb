```python
%%bash
mkdir -p data/fastq data/ref_genome results/sai results/sam results/bam results/bcf results/vcf
```

```python
%%bash
mkdir -p results/sai
mkdir -p results/sam
mkdir -p results/bam
mkdir -p results/bcf
mkdir -p results/vcf
mkdir -p data/ref_genome
```

```python
%%bash
wget https://hgdownload.soe.ucsc.edu/goldenPath/hg38/chromosomes/chr17.fa.gz -P data/ref_genome #unduh chr17
```

```python
%%bash
gzip -d  data/ref_genome/chr17.fa.gz #unzip .fa.gz menjadi .fa
samtools faidx data/ref_genome/chr17.fa #buat indeks (.fa.fai)
```

```python
%%bash
bwa index -p data/ref_genome/chr17bwaidx -a bwtsw data/ref_genome/chr17.fa #indeksasi genom
```

```python
%%bash
#Sampel 1
bwa aln -t 4 data/ref_genome/chr17bwaidx data/fastq/DRR491202_1.fastq > DRR491202_1.sai #align bacaan forward reads dri sampel 1
bwa aln -t 4 data/ref_genome/chr17bwaidx data/fastq/DRR491202_2.fastq > DRR491202_2.sai #align bacaan reverse reads dri sampel 1
mv *.sai results/sai/ #pindah output .sai ke folder result
```

```python
%%bash
#Sampel 2
bwa aln -t 4 data/ref_genome/chr17bwaidx data/fastq/DRR491206_1.fastq > DRR491206_1.sai
bwa aln -t 4 data/ref_genome/chr17bwaidx data/fastq/DRR491206_2.fastq > DRR491206_2.sai
mv *.sai results/sai/
```

```python
%%bash
#Sampel 1
bwa sampe data/ref_genome/chr17bwaidx results/sai/DRR491202_1.sai results/sai/DRR491202_2.sai data/fastq/DRR491202_1.fastq data/fastq/DRR491202_2.fastq > DRR491202.sam #konversi ke SAM
mv *.sam results/sam #pindah output .sam ke folder result
```

```python
%%bash
#Sampel 2
bwa sampe data/ref_genome/chr17bwaidx results/sai/DRR491206_1.sai results/sai/DRR491206_2.sai data/fastq/DRR491206_1.fastq data/fastq/DRR491206_2.fastq > DRR491206.sam #konversi ke SAM
mv *.sam results/sam #pindah output .sam ke folder result
```

```python
%%bash
cat results/sam/DRR491202.sam | head
```

```python
%%bash
cat results/sam/DRR491206.sam | head
```

```python
%%bash
#Sampel 1
samtools view -bS results/sam/DRR491202.sam > DRR491202.bam #konversi SAM ke BAM
samtools sort DRR491202.bam -o DRR491202_sorted.bam #sort BAM file
samtools index DRR491202_sorted.bam #buat indeks BAM
mv *.bam *.bai results/bam/ #pindah output .bam dan .bai ke folder result
```

```python
%%bash
#Sampel 2
samtools view -bS results/sam/DRR491206.sam > DRR491206.bam #konversi SAM ke BAM
samtools sort DRR491206.bam -o DRR491206_sorted.bam #sort BAM file
samtools index DRR491206_sorted.bam #buat indeks BAM
mv *.bam *.bai results/bam/ #pindah output .bam dan .bai ke folder result
```

```python
%%bash
#Sampel 1
bcftools mpileup -O b -o DRR491202_raw.bcf -f data/ref_genome/chr17.fa results/bam/DRR491202_sorted.bam #hitung read coverage dan tandai variasi
mv *.bcf results/bcf #pindah output .bcf ke folder result
```

```python
%%bash
#Sampel 2
bcftools mpileup -O b -o DRR491206_raw.bcf -f data/ref_genome/chr17.fa results/bam/DRR491206_sorted.bam #hitung read coverage dan tandai variasi
mv *.bcf results/bcf #pindah output .bcf ke folder result
```

```python
%%bash
#Sampel 1
bcftools call --ploidy 1 -m -v -o DRR491202_variants.vcf results/bcf/DRR491202_raw.bcf #pemanggilan variasi
mv *.vcf results/vcf #pindah output .vcf ke folder result
```

```python
%%bash
#Sampel 1
bcftools call --ploidy 1 -m -v -o DRR491206_variants.vcf results/bcf/DRR491206_raw.bcf #pemanggilan variasi
mv *.vcf results/vcf #pindah output .vcf ke folder result
```

```python
%%bash
#Sampel 1
vcfutils.pl varFilter results/vcf/DRR491202_variants.vcf  > DRR491202_final_variants.vcf #pemanggilan variasi
mv *.vcf results/vcf #pindah output .vcf ke folder result
```

```python
%%bash
#Sampel 2
vcfutils.pl varFilter results/vcf/DRR491206_variants.vcf  > DRR491206_final_variants.vcf #pemanggilan variasi
mv *.vcf results/vcf #pindah output .vcf ke folder result
```
