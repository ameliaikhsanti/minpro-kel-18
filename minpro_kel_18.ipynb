```python
%%bash
mkdir -p results/sai
mkdir -p results/sam
mkdir -p results/bam
mkdir -p results/bcf
mkdir -p results/vcf
mkdir -p data/ref_genome
```

```python
%%bash
wget https://hgdownload.soe.ucsc.edu/goldenPath/hg38/chromosomes/chr17.fa.gz -P data/ref_genome #unduh chr17
```

```python
%%bash
wget -P data/fastq/ "https://ftp.sra.ebi.ac.uk/vol1/fastq/DRR491/DRR491202/DRR491202_1.fastq.gz"
wget -P data/fastq/ "https://ftp.sra.ebi.ac.uk/vol1/fastq/DRR491/DRR491202/DRR491202_2.fastq.gz"
wget -P data/fastq/ "https://ftp.sra.ebi.ac.uk/vol1/fastq/DRR491/DRR491206/DRR491206_1.fastq.gz"
wget -P data/fastq/ "https://ftp.sra.ebi.ac.uk/vol1/fastq/DRR491/DRR491206/DRR491206_2.fastq.gz"
wget -P data/fastq/ "https://ftp.sra.ebi.ac.uk/vol1/fastq/DRR491/DRR491212/DRR491212_1.fastq.gz"
wget -P data/fastq/ "https://ftp.sra.ebi.ac.uk/vol1/fastq/DRR491/DRR491212/DRR491212_2.fastq.gz"
wget -P data/fastq/ "https://ftp.sra.ebi.ac.uk/vol1/fastq/DRR491/DRR491220/DRR491220_1.fastq.gz"
wget -P data/fastq/ "https://ftp.sra.ebi.ac.uk/vol1/fastq/DRR491/DRR491220/DRR491220_2.fastq.gz"
wget -P data/fastq/ "https://ftp.sra.ebi.ac.uk/vol1/fastq/DRR491/DRR491225/DRR491225_1.fastq.gz"
wget -P data/fastq/ "https://ftp.sra.ebi.ac.uk/vol1/fastq/DRR491/DRR491225/DRR491225_2.fastq.gz"
wget -P data/fastq/ "https://ftp.sra.ebi.ac.uk/vol1/fastq/DRR491/DRR491245/DRR491245_1.fastq.gz"
wget -P data/fastq/ "https://ftp.sra.ebi.ac.uk/vol1/fastq/DRR491/DRR491245/DRR491245_2.fastq.gz"
```

```python
%%bash
gunzip data/fastq/DRR491202_1.fastq.gz
gunzip data/fastq/DRR491202_2.fastq.gz
gunzip data/fastq/DRR491206_1.fastq.gz
gunzip data/fastq/DRR491206_2.fastq.gz
gunzip data/fastq/DRR491212_1.fastq.gz
gunzip data/fastq/DRR491212_2.fastq.gz
gunzip data/fastq/DRR491220_1.fastq.gz
gunzip data/fastq/DRR491220_2.fastq.gz
gunzip data/fastq/DRR491225_1.fastq.gz
gunzip data/fastq/DRR491225_2.fastq.gz
gunzip data/fastq/DRR491245_1.fastq.gz
gunzip data/fastq/DRR491245_2.fastq.gz
```

```python
%%bash
gzip -d  data/ref_genome/chr17.fa.gz #unzip .fa.gz menjadi .fa
samtools faidx data/ref_genome/chr17.fa #buat indeks (.fa.fai)
```

```python
%%bash
bwa index -p data/ref_genome/chr17bwaidx -a bwtsw data/ref_genome/chr17.fa #indeksasi genom
```

```python
%%bash
for sample in DRR491202 DRR491206 DRR491212 DRR491220 DRR491225 DRR491245; do
    bwa aln -t 4 data/ref_genome/chr17bwaidx data/fastq/${sample}_1.fastq > results/sai/${sample}_1.sai
    bwa aln -t 4 data/ref_genome/chr17bwaidx data/fastq/${sample}_2.fastq > results/sai/${sample}_2.sai
done
```

```python
%%bash
for sample in DRR491202 DRR491206 DRR491212 DRR491220 DRR491225 DRR491245; do
bwa sampe data/ref_genome/chr17bwaidx results/sai/${sample}_1.sai results/sai/${sample}_2.sai data/fastq/${sample}_1.fastq data/fastq/${sample}_2.fastq > results/sam/${sample}.sam
done
```

```python
%%bash
for sample in DRR491202 DRR491206 DRR491212 DRR491220 DRR491225 DRR491245; do
cat results/sam/${sample}.sam | head
done
```

```python
%%bash
for sample in DRR491202 DRR491206 DRR491212 DRR491220 DRR491225 DRR491245; do
samtools view -bS results/sam/${sample}.sam > ${sample}.bam #konversi SAM ke BAM
samtools sort ${sample}.bam -o ${sample}_sorted.bam #sort BAM file
samtools index ${sample}_sorted.bam #buat indeks BAM
mv *.bam *.bai results/bam/ #pindah output .bam dan .bai ke folder result
done
```

```python
%%bash
for sample in DRR491202 DRR491206 DRR491212 DRR491220 DRR491225 DRR491245; do
bcftools mpileup -O b -o ${sample}_raw.bcf -f data/ref_genome/chr17.fa results/bam/${sample}_sorted.bam #hitung read coverage dan tandai variasi
mv *.bcf results/bcf #pindah output .bcf ke folder result
done
```

```python
%%bash
for sample in DRR491202 DRR491206 DRR491212 DRR491220 DRR491225 DRR491245; do
bcftools call --ploidy 1 -m -v -o ${sample}_variants.vcf results/bcf/${sample}_raw.bcf #pemanggilan variasi
mv *.vcf results/vcf #pindah output .vcf ke folder result
done
```

```python
%%bash
for sample in DRR491202 DRR491206 DRR491212 DRR491220 DRR491225 DRR491245; do
vcfutils.pl varFilter results/vcf/${sample}_variants.vcf > results/vcf/${sample}_final_variants.vcf #pemanggilan variasi
done
```

```python
%%bash
for sample in DRR491202 DRR491206 DRR491212 DRR491220 DRR491225 DRR491245; do
cat results/vcf/${sample}_final_variants.vcf | grep -A10 '#CHROM'
done
```
